```python
ego_url = 'https://raw.githubusercontent.com/p19pago/semantics-2023/main/project/project%20files/ego.csv'
```

```python
ego_frame = pd.read_csv(ego_url)
```

```python
import networkx as nx
```

```python
import numpy as np
```

```python
G = nx.Graph()
```

```python
for index, row in ego_frame.iterrows():
    source = row['source']
    target = row['target']
    G.add_edge(source, target)
```

```python
edge_list = [
              ['source', 'target'],
              ['source', 'target']
            ]
```

```python
#graph stats

G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
# graph based metrics
```

```python
nx.average_shortest_path_length(G)
```

```python
nx.average_clustering(G)
```

```python
degrees = G.degree()
```

```python
#degrees is a dict, get values:

degrees_num = [item[1] for item in degrees]
mean_degree = np.mean(degrees_num)
```

```python
print(degrees_num)
```

```python
print(mean_degree)
```

```python
# οπτικοποίηση γράφου
```

```python
import matplotlib.pyplot as plt
```

```python
pos = nx.spring_layout(G) #specify layout for visual

plt.figure()

nodes = nx.draw_networkx_nodes(G, pos, alpha=0.8)
nodes.set_edgecolor('k') #black
nx.draw_networkx_labels(G, pos, font_size=8)
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.2)
plt.show()
```

```python
usersfriends_url = 'https://raw.githubusercontent.com/p19pago/semantics-2023/main/project/project%20files/users_friends.csv'
```

```python
uf_frame = pd.read_csv(usersfriends_url)
```

```python
GF = nx.Graph()
```

```python
for index, row in uf_frame.iterrows():
  id_a = row['id_a']
  id_b = row['id_b']
  GF.add_edge(id_a, id_b)
```

```python
edgelist_GF = [ ['id_a', 'id_b'],
                ['id_a', 'id_b']
]
```

```python
# graph stats
```

```python
GF.number_of_nodes()
```

```python
GF.number_of_edges()
```

```python
GF.add_edges_from(edgelist_GF)
```

```python
# Υπολογίστε τον αριθμό των συνδεδεμένων συνιστωσών

num_components = nx.number_connected_components(GF)
print("Number of connected components is: ", num_components)
```

```python
# μετρικές με βάση τους γράφους
```

```python
# Χαρακτηριστικό μήκος μονοπατιού για κάθε συνδεδεμένη συνιστώσα, ξεχωριστά

# χρήση μεθόδου υπογράφων (subgraphs)

for num_components in nx.connected_components(GF):
    subgraph = GF.subgraph(num_components)
    if subgraph.number_of_nodes() > 1:
        characteristic_path_length = nx.average_shortest_path_length(subgraph)
        print("Το χαρακτηριστικό μήκος μονοπατιού για την συνιστώσα", num_components, "είναι:", characteristic_path_length)
```

```python
# Μέσος βαθμός του γράφου

degrees_GF = GF.degree()
```

```python
print(degrees_GF)
```

```python
# degrees_GF is a dict, get values:

degrees_num_GF = [item[1] for item in degrees_GF]
mean_degree_GF = np.mean(degrees_num_GF)
```

```python
print(degrees_num_GF)
```

```python
nx.average_clustering(GF)
```
